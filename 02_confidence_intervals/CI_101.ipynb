# 1. What is a Confidence Interval?

A Confidence Interval is a **range of values** that **likely contains** the **true value of a population parameter** (e.g., mean or proportion).

It tells us how uncertain we are about our estimate from a sample.

The confidence level (e.g., 95%) shows how sure we are that the true population value falls within this range.

Example in Real Life:

Polls reporting “Candidate X has 48% support ± 3%” are giving a confidence interval.

A 95% CI means that if we repeated the poll many times, 95% of the intervals calculated would contain the true support.

## 2. How Samples and Populations Relate to CI

- Population: All individuals or data points we care about (e.g., all customers, all students).
- Sample: A smaller group we actually measure, used to infer the population.

Why we sample:

- Collecting data from the entire population is often too costly or impossible.

- Samples allow us to make estimates about the population efficiently.

Important: Because the sample may not perfectly represent the population, we quantify the uncertainty using a CI.

## 3. Sample Mean vs Population Mean

| Term                     | Description                            | Example                           |
| ------------------------ | -------------------------------------- | --------------------------------- |
| Population Mean (μ)      | True average of the population         | All student heights = 170 cm      |
| Sample Mean (x̄)         | Average of a sample used to estimate μ | 30 students’ mean height = 171 cm |
| Confidence Interval (CI) | Range likely to contain μ              | 168 cm – 172 cm, 95% confidence   |


### Key Takeaways
Population: The entire group you're interested in studying.

Sample: A smaller group selected from the population, used to make inferences about the population.


Confidence Interval (CI): A range of values, derived from the sample, that is likely to contain the population parameter (e.g., mean), along with a confidence level (e.g., 95%) indicating the certainty of that range.

Key idea: CI provides a range around the sample mean, showing where the population mean likely lies.

### Example: Picking up where we left off in the Smaple and Population notebook

We will generate the data again and now we will calculate the CI for the given sample data.

Reminder this is a simulation to showcase why we need CI when dealing with sample data. 

If you have access to the population data and no other constraints.

Then use the population data. But this not always the case



````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Population: all student heights
np.random.seed(0) # this ensures we can always re-run this block of code and have the same results otherwise it would generate a differnt sample on each run
population = np.random.normal(170, 10, 5000)

# Sample of 30 students
sample = np.random.choice(population, 30, replace=False)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1) # investigate what the ddof parameter does...
````

#### Note:

We will calculate the CI at 95%. 
Forget that we have access to the population data, we only know how much data we could have i.e. 5000.

Everything we do now we will do on the sample data

But before we can do this, ask yourself:

- What Data Type am I dealing with? Continuous or Discrete
- What Distribution is this likely going to be? eg. Normal, Exponential, Poisson etc...

This is the key in calculating a CI that is actually representative of the data.

Remember our data has only 30 samples. This might result in messy/noisy distribution purely because we have few data points.
In such cases plot it nonetheless but remember:
- the histogram of the sample is just an estimate of the population distribution.

- the CI calculation doesn’t require the sample histogram to look perfectly normal — it relies on the assumption that the underlying population is normal (or that the Central Limit Theorem allows approximation for large enough n).

Don’t be thrown off by sample noise; the key is the population assumption and the size of your sample.

````python
# 95% CI using Z-score (approximation for illustration) Detailed code explanation at the end of NB

z = norm.ppf(0.975)  # percent point function, which is the inverse of the CDF (cumulative distribution function).
margin_error = z * (sample_std / np.sqrt(len(sample))) # how far above or below the sample mean we go to capture the true mean with 95% confidence.
ci_lower = sample_mean - margin_error # Lower bound of the confidence interval.
ci_upper = sample_mean + margin_error # Upper bound of the confidence interval.

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% CI: ({ci_lower:.2f}, {ci_upper:.2f})")
````

#### Detailed explanation of the CI calcultation: 

```python
z = norm.ppf(0.975)  # percent point function, which is the inverse of the CDF (cumulative distribution function).
margin_error = z * (sample_std / np.sqrt(len(sample))) # how far above or below the sample mean we go to capture the true mean with 95% confidence.
ci_lower = sample_mean - margin_error # Lower bound of the confidence interval.
ci_upper = sample_mean + margin_error # Upper bound of the confidence interval.

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% CI: ({ci_lower:.2f}, {ci_upper:.2f})")
```





`z = norm.ppf(0.975)`

- `norm.ppf()` = percent point function, which is the inverse of the CDF (cumulative distribution function). Remember the CDF id the function that gives you the probabilities up to a certain point. We know the probability already what we want is the corresponding value. Step in the PPF.

- Here we ask: “What z-score corresponds to the point where 97.5% of the area under the normal curve is to the left?”

- For a 95% CI, we split the 5% “error” evenly: 2.5% in each tail → so we want the cutoff at 97.5%. Imagine a normal disttribution curve and cut the tails at 2.5% and 97.5%. You remain with 95% in the middle.

- Result: z ≈ 1.96, the critical value used in 95% CI formulas.

- <img src="./images/CI.jpg" alt="CI" width="600"/>

---- 

`margin_error = z * (sample_std / np.sqrt(len(sample)))`

- `sample_std` = standard deviation of our sample (how spread out the data is).

- `len(sample)` = number of observations in the sample.

- `np.sqrt(len(sample)) = √n`, which gives the scaling for standard error. Larger samples → smaller error.

- `(sample_std / np.sqrt(len(sample)))` = standard error (SE), i.e., how much the sample mean might vary from the population mean.

- Multiplying SE by z gives the margin of error → how far above or below the sample mean we go to capture the true mean with 95% confidence.

### What is a 95% Confidence Interval?

A 95% confidence interval is a range of values that you can be 95% certain contains the true population mean. It is calculated using the following formula:


````python
from IPython.display import display, Math

# Display LaTeX formula for confidence interval
display(Math(r'\text{CI} = \left(\bar{x} - Z \times \frac{\sigma}{\sqrt{n}}, \bar{x} + Z \times \frac{\sigma}{\sqrt{n}}\right)'))

# Display additional text for definitions with explicit newline characters
display(Math(r'\text{Where:} \newline \bar{x} \text{ is the sample mean} \newline Z \text{ is the Z-score corresponding to the desired confidence level (e.g., 1.96 for 95\%)} \newline \sigma \text{ is the population standard deviation} \newline n \text{ is the sample size}'))
````
