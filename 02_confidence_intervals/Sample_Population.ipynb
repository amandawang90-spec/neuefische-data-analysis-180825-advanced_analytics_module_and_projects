## 2. Samples

    Definition:
    
    A sample is a subset of the population that is selected for analysis.

    Why we sample:

        Measuring the entire population is often too costly, too time-consuming, or impossible.

        A well-chosen sample can give accurate estimates about the population.

----

### Example

We will generate some data about students who have an average height of 170cm and a standard deviation fo 10cm.

This will be our population data whereby the average can be thought of as the TRUE average.

We will then sample this data and note the difference in the average, if any.


```python
import numpy as np
import matplotlib.pyplot as plt

# Population: all student heights
np.random.seed(0) # this ensures we can always re-run this block of code and have the same results otherwise it would generate a differnt sample on each run
population = np.random.normal(170, 10, 5000)  # mean=170 cm, sd=10 cm

plt.hist(population, bins=30, alpha=0.5, color='orange')
plt.title("Population: All 5000 students")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
#Random sample of 30 students

sample = np.random.choice(population, size=30, replace=False) 
plt.hist(sample, bins=10, alpha=0.7, color='blue')
plt.title("Sample: 30 students")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.show()

print("Sample mean:", np.mean(sample))
print("Population mean:", np.mean(population))
```

## 3. Sample Statistics vs Population Parameters 



| Concept             | Definition                     | Example                        |
|--------------------|--------------------------------|--------------------------------|
| Population Parameter | True value describing the population | Population mean height = 170 cm |
| Sample Statistic     | Estimate from the sample       | Sample mean height = 171.1 cm |


## 4. Why Sampling Matters in Analytics

    Enables cost-effective analysis.

    Allows us to make predictions and decisions about the population.

    Introduces uncertainty, which is later quantified using Confidence Intervals.



### Mini Case Example:

## 5. Takeaways

    - Population = complete set of data points we care about.

    - Sample = a smaller, manageable subset used to estimate the population.

    - Sample statistics estimate population parameters.

    - Sampling introduces uncertainty → the next step is to quantify it with Confidence Intervals.
