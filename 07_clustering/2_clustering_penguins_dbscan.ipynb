#### Determining DBSCAN hyperparameters
What do we need as arguments for DBSCAN(eps=?, min_samples=?)

**min_samples**  
The min_samples might be iteratively refined based on experimental results and insights gained from evaluating clustering performance metrics. We start with min_samples = 4 (as the count of our features).

**eps**  
The elbow plot method can be adapted for DBSCAN to help determine an appropriate value for the eps (epsilon) parameter. Here's a step-by-step description of how to use this method:

1. Compute the k-distances:
For each data point, compute the distance to its k-th nearest neighbor, where k is typically set to the value of min_samples.
2. Sort the distances:
Sort these k-distances in ascending order.
3. Plot the distances:
Create a plot of these sorted k-distances against their indices.
4. Identify the "elbow" point:
Look for a point in the plot where the slope of the curve changes sharply, resembling an "elbow". This point is often considered a good candidate for eps.
The elbow point represents the distance at which the density of points changes significantly, helping to distinguish between points that are close enough to be considered part of the same cluster and points that are further apart.

```python
from sklearn.neighbors import NearestNeighbors

X = df_num_scaled

# # Fit nearest neighbors
# Initializes a NearestNeighbors object with the parameter n_neighbors set to 4. 
# This means that for each point in your dataset X, it will find the 4 nearest neighbors.
nbrs = NearestNeighbors(n_neighbors=4).fit(X)

# 'distances': Array where each element distances[i] contains the distances to the 4 nearest neighbors of point X[i].
# '_': Array of indices of the 4 nearest neighbors of each point X[i]. Not used in the following code.
distances, _ = nbrs.kneighbors(X)
```

```python
# Sort distances and plot
distances = np.sort(distances[:, 1], axis=0)  # Sort by distance to the second nearest neighbor

# Plot the elbow
plt.figure(figsize=(10, 6))
plt.plot(distances)
plt.xlabel('Points sorted by distance')
plt.ylabel('Distance to nearest neighbor')
plt.title('Elbow Plot for DBSCAN hyperparameter epsilon')
plt.grid(True)
plt.show()
```

The sorted distances are plotted, with the x-axis representing the sorted data points and the y-axis representing the k-distances.
Interpreting the Elbow Plot:
The "elbow" point in the plot, where the curve bends and the slope changes sharply, indicates a suitable value for eps. This point marks a transition from points that are close together (and likely in the same cluster) to points that are further apart.  

We might choose eps=0.65, based on our findings.  

By using the elbow plot method, you can visually determine an appropriate value for eps that helps DBSCAN distinguish between dense regions (clusters) and sparser regions (noise) in your dataset.

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.65
                ,min_samples=5)
dbscan.fit(df_num_scaled)
```
